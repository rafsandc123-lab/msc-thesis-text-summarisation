```python
# =====================================================
# FULL DATASET TEXT CLASSIFICATION (THESIS-READY VERSION)
# IMDB DATASET
# BERT + XLNET
# NO SUMMARISATION -> CLASSIFICATION
# WITH:
# - Accuracy / Precision / Recall / F1 / ROC-AUC
# - Confusion Matrix / Classification Report / ROC Curve
# - Training vs Validation Loss Curves
# - Training Time / Inference Latency / Throughput
# - GPU Memory Usage
# - Token Statistics / Truncation Statistics
# =====================================================

!pip install -q pandas numpy matplotlib scikit-learn transformers datasets sentencepiece accelerate torch

import os
import re
import time
import random
import warnings
import numpy as np
import pandas as pd
import torch
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
    roc_curve,
    auc
)

from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
    DataCollatorWithPadding,
    set_seed
)

warnings.filterwarnings("ignore")

# -------------------------
# REPRODUCIBILITY
# -------------------------
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)
set_seed(SEED)

print("Torch version:", torch.__version__)
print("CUDA available:", torch.cuda.is_available())
device_name = torch.cuda.get_device_name(0) if torch.cuda.is_available() else "CPU"
print("Device:", device_name)

# -------------------------
# PATHS
# -------------------------
DATA_PATH = "/content/IMDB Dataset.csv"
BASE_OUTPUT_DIR = "/content/thesis_outputs"
os.makedirs(BASE_OUTPUT_DIR, exist_ok=True)

# -------------------------
# LOAD DATASET
# -------------------------
df = pd.read_csv(DATA_PATH)

print("Original dataset shape:", df.shape)
display(df.head())

# -------------------------
# CLEAN TEXT
# -------------------------
def clean_text(text):
    text = str(text)
    text = re.sub(r"<.*?>", " ", text)   # remove HTML
    text = re.sub(r"http\S+", "", text)  # remove URLs
    text = re.sub(r"\s+", " ", text)     # normalize spaces
    return text.strip()

df = df.dropna(subset=["review", "sentiment"]).drop_duplicates().reset_index(drop=True)
df["clean_text"] = df["review"].apply(clean_text)

label_map = {"positive": 1, "negative": 0}
df["label"] = df["sentiment"].map(label_map)

print("Cleaned dataset shape:", df.shape)
print("\nLabel distribution:")
print(df["label"].value_counts())

# -------------------------
# TRAIN / VAL / TEST SPLIT
# -------------------------
train_texts, temp_texts, train_labels, temp_labels = train_test_split(
    df["clean_text"],
    df["label"],
    test_size=0.30,
    stratify=df["label"],
    random_state=SEED
)

val_texts, test_texts, val_labels, test_labels = train_test_split(
    temp_texts,
    temp_labels,
    test_size=0.50,
    stratify=temp_labels,
    random_state=SEED
)

train_texts = train_texts.tolist()
val_texts = val_texts.tolist()
test_texts = test_texts.tolist()

train_labels = train_labels.tolist()
val_labels = val_labels.tolist()
test_labels = test_labels.tolist()

print("\nSplit sizes")
print("Train:", len(train_texts))
print("Validation:", len(val_texts))
print("Test:", len(test_texts))

# -------------------------
# DATASET CLASS
# -------------------------
class TextDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {k: torch.tensor(v[idx]) for k, v in self.encodings.items()}
        item["labels"] = torch.tensor(self.labels[idx], dtype=torch.long)
        return item

    def __len__(self):
        return len(self.labels)

# -------------------------
# METRICS FUNCTION
# -------------------------
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=1)

    return {
        "accuracy": accuracy_score(labels, preds),
        "precision": precision_score(labels, preds, zero_division=0),
        "recall": recall_score(labels, preds, zero_division=0),
        "f1": f1_score(labels, preds, zero_division=0)
    }

# -------------------------
# HELPER FUNCTIONS
# -------------------------
def bytes_to_gb(num_bytes):
    return num_bytes / (1024 ** 3)

def summarize_lengths(lengths):
    lengths = np.array(lengths)
    return {
        "mean": float(np.mean(lengths)),
        "median": float(np.median(lengths)),
        "min": int(np.min(lengths)),
        "max": int(np.max(lengths)),
        "p95": float(np.percentile(lengths, 95))
    }

def get_token_stats(tokenizer, texts, max_length, split_name="Split"):
    original_lengths = []
    truncated_lengths = []
    truncated_count = 0

    for text in texts:
        # Raw length without truncation
        raw_tokens = tokenizer.tokenize(text)
        orig_len = len(raw_tokens) + 2  # approximate special tokens

        # Length after truncation
        trunc_enc = tokenizer(
            text,
            add_special_tokens=True,
            truncation=True,
            max_length=max_length,
            padding=False
        )
        trunc_len = len(trunc_enc["input_ids"])

        original_lengths.append(orig_len)
        truncated_lengths.append(trunc_len)

        if orig_len > max_length:
            truncated_count += 1

    original_stats = summarize_lengths(original_lengths)
    truncated_stats = summarize_lengths(truncated_lengths)
    trunc_pct = (truncated_count / len(texts)) * 100

    return {
        "split": split_name,
        "n_samples": len(texts),
        "max_length": max_length,
        "original_mean_tokens": round(original_stats["mean"], 2),
        "original_median_tokens": round(original_stats["median"], 2),
        "original_min_tokens": original_stats["min"],
        "original_max_tokens": original_stats["max"],
        "original_p95_tokens": round(original_stats["p95"], 2),
        "truncated_mean_tokens": round(truncated_stats["mean"], 2),
        "truncated_median_tokens": round(truncated_stats["median"], 2),
        "truncated_min_tokens": truncated_stats["min"],
        "truncated_max_tokens": truncated_stats["max"],
        "truncated_p95_tokens": round(truncated_stats["p95"], 2),
        "num_truncated_samples": truncated_count,
        "truncation_percentage": round(trunc_pct, 2)
    }

def print_token_stats(model_name, stats_list):
    print(f"\n=== {model_name} TOKEN & TRUNCATION STATISTICS ===")
    for stats in stats_list:
        print(f"\n[{stats['split']}]")
        print(f"Samples: {stats['n_samples']}")
        print(f"Max length: {stats['max_length']}")
        print(f"Original mean/median tokens: {stats['original_mean_tokens']} / {stats['original_median_tokens']}")
        print(f"Original min/max tokens: {stats['original_min_tokens']} / {stats['original_max_tokens']}")
        print(f"Original 95th percentile: {stats['original_p95_tokens']}")
        print(f"Post-truncation mean/median tokens: {stats['truncated_mean_tokens']} / {stats['truncated_median_tokens']}")
        print(f"Post-truncation min/max tokens: {stats['truncated_min_tokens']} / {stats['truncated_max_tokens']}")
        print(f"Post-truncation 95th percentile: {stats['truncated_p95_tokens']}")
        print(f"Truncated samples: {stats['num_truncated_samples']}")
        print(f"Truncation percentage: {stats['truncation_percentage']}%")

def extract_loss_history(trainer):
    logs = trainer.state.log_history

    train_records = []
    eval_records = []

    for log in logs:
        if "loss" in log and "epoch" in log and "eval_loss" not in log:
            train_records.append((float(log["epoch"]), float(log["loss"])))
        if "eval_loss" in log and "epoch" in log:
            eval_records.append((float(log["epoch"]), float(log["eval_loss"])))

    train_dict = {}
    for ep, loss in train_records:
        train_dict[ep] = loss

    eval_dict = {}
    for ep, loss in eval_records:
        eval_dict[ep] = loss

    train_epochs = sorted(train_dict.keys())
    eval_epochs = sorted(eval_dict.keys())

    train_losses = [train_dict[e] for e in train_epochs]
    eval_losses = [eval_dict[e] for e in eval_epochs]

    return train_epochs, train_losses, eval_epochs, eval_losses

def plot_loss_curves(trainer, model_name):
    train_epochs, train_losses, eval_epochs, eval_losses = extract_loss_history(trainer)

    plt.figure(figsize=(6, 4))
    if len(train_epochs) > 0:
        plt.plot(train_epochs, train_losses, marker="o", label="Training Loss")
    if len(eval_epochs) > 0:
        plt.plot(eval_epochs, eval_losses, marker="o", label="Validation Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title(f"Training vs Validation Loss - {model_name}")
    plt.legend()
    plt.grid(True)
    plt.show()

def reset_gpu_memory():
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        torch.cuda.reset_peak_memory_stats()

def get_gpu_memory_stats():
    if torch.cuda.is_available():
        allocated = torch.cuda.max_memory_allocated()
        reserved = torch.cuda.max_memory_reserved()
        return {
            "peak_gpu_memory_allocated_gb": round(bytes_to_gb(allocated), 4),
            "peak_gpu_memory_reserved_gb": round(bytes_to_gb(reserved), 4)
        }
    else:
        return {
            "peak_gpu_memory_allocated_gb": 0.0,
            "peak_gpu_memory_reserved_gb": 0.0
        }

def evaluate_predictions(y_true, logits):
    y_pred = np.argmax(logits, axis=1)
    y_prob = torch.softmax(torch.tensor(logits), dim=1)[:, 1].numpy()

    metrics = {
        "accuracy": accuracy_score(y_true, y_pred),
        "precision": precision_score(y_true, y_pred, zero_division=0),
        "recall": recall_score(y_true, y_pred, zero_division=0),
        "f1": f1_score(y_true, y_pred, zero_division=0),
        "roc_auc": roc_auc_score(y_true, y_prob)
    }

    return y_pred, y_prob, metrics

def run_experiment(
    model_label,
    model_name,
    train_texts,
    val_texts,
    test_texts,
    train_labels,
    val_labels,
    test_labels,
    max_length=256,
    batch_size=8,
    epochs=2,
    learning_rate=2e-5,
    weight_decay=0.01
):
    print("\n" + "=" * 60)
    print(f"RUNNING: {model_label}")
    print("=" * 60)

    output_dir = os.path.join(BASE_OUTPUT_DIR, f"{model_label.lower()}_results")
    os.makedirs(output_dir, exist_ok=True)

    # -------------------------
    # TOKENIZER
    # -------------------------
    tokenizer = AutoTokenizer.from_pretrained(model_name)

    if tokenizer.pad_token is None:
        tokenizer.pad_token = tokenizer.eos_token if tokenizer.eos_token is not None else tokenizer.sep_token

    # -------------------------
    # TOKEN / TRUNCATION STATS
    # -------------------------
    train_token_stats = get_token_stats(tokenizer, train_texts, max_length=max_length, split_name="Train")
    val_token_stats = get_token_stats(tokenizer, val_texts, max_length=max_length, split_name="Validation")
    test_token_stats = get_token_stats(tokenizer, test_texts, max_length=max_length, split_name="Test")

    print_token_stats(model_label, [train_token_stats, val_token_stats, test_token_stats])

    # -------------------------
    # ENCODING
    # -------------------------
    train_enc = tokenizer(train_texts, truncation=True, padding=False, max_length=max_length)
    val_enc = tokenizer(val_texts, truncation=True, padding=False, max_length=max_length)
    test_enc = tokenizer(test_texts, truncation=True, padding=False, max_length=max_length)

    train_dataset = TextDataset(train_enc, train_labels)
    val_dataset = TextDataset(val_enc, val_labels)
    test_dataset = TextDataset(test_enc, test_labels)

    data_collator = DataCollatorWithPadding(tokenizer=tokenizer)

    # -------------------------
    # MODEL
    # -------------------------
    model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)
    model.config.pad_token_id = tokenizer.pad_token_id

    # -------------------------
    # TRAINING ARGUMENTS
    # -------------------------
    args = TrainingArguments(
        output_dir=output_dir,
        eval_strategy="epoch",
        save_strategy="epoch",
        logging_strategy="epoch",
        save_total_limit=2,
        per_device_train_batch_size=batch_size,
        per_device_eval_batch_size=batch_size,
        num_train_epochs=epochs,
        learning_rate=learning_rate,
        weight_decay=weight_decay,
        load_best_model_at_end=True,
        metric_for_best_model="f1",
        greater_is_better=True,
        fp16=torch.cuda.is_available(),
        report_to="none",
        seed=SEED
    )

    trainer = Trainer(
        model=model,
        args=args,
        train_dataset=train_dataset,
        eval_dataset=val_dataset,
        compute_metrics=compute_metrics,
        data_collator=data_collator
    )

    # -------------------------
    # TRAINING
    # -------------------------
    print(f"\nTraining {model_label}...")

    reset_gpu_memory()
    train_start = time.time()
    trainer.train()
    train_time = time.time() - train_start
    train_mem_stats = get_gpu_memory_stats()

    train_samples_per_sec = len(train_dataset) / train_time
    train_effective_tokens = sum(len(x) for x in train_enc["input_ids"])
    train_tokens_per_sec = train_effective_tokens / train_time

    # -------------------------
    # INFERENCE
    # -------------------------
    print(f"\nEvaluating {model_label} on test set...")

    reset_gpu_memory()
    infer_start = time.time()
    pred_output = trainer.predict(test_dataset)
    infer_time = time.time() - infer_start
    infer_mem_stats = get_gpu_memory_stats()

    y_true = np.array(test_labels)
    y_pred, y_prob, metrics = evaluate_predictions(y_true, pred_output.predictions)

    inference_samples_per_sec = len(test_dataset) / infer_time
    inference_latency_per_sample_ms = (infer_time / len(test_dataset)) * 1000
    test_effective_tokens = sum(len(x) for x in test_enc["input_ids"])
    inference_tokens_per_sec = test_effective_tokens / infer_time

    # -------------------------
    # RESULTS SUMMARY
    # -------------------------
    print(f"\n=== {model_label} RESULTS ===")
    print("Accuracy:", round(metrics["accuracy"], 4))
    print("Precision:", round(metrics["precision"], 4))
    print("Recall:", round(metrics["recall"], 4))
    print("F1:", round(metrics["f1"], 4))
    print("ROC-AUC:", round(metrics["roc_auc"], 4))
    print("Training time (min):", round(train_time / 60, 2))
    print("Inference time (sec):", round(infer_time, 2))
    print("Inference latency per sample (ms):", round(inference_latency_per_sample_ms, 4))
    print("Training throughput (samples/sec):", round(train_samples_per_sec, 2))
    print("Training throughput (tokens/sec):", round(train_tokens_per_sec, 2))
    print("Inference throughput (samples/sec):", round(inference_samples_per_sec, 2))
    print("Inference throughput (tokens/sec):", round(inference_tokens_per_sec, 2))
    print("Peak GPU memory allocated during training (GB):", train_mem_stats["peak_gpu_memory_allocated_gb"])
    print("Peak GPU memory reserved during training (GB):", train_mem_stats["peak_gpu_memory_reserved_gb"])
    print("Peak GPU memory allocated during inference (GB):", infer_mem_stats["peak_gpu_memory_allocated_gb"])
    print("Peak GPU memory reserved during inference (GB):", infer_mem_stats["peak_gpu_memory_reserved_gb"])

    # -------------------------
    # CLASSIFICATION REPORT
    # -------------------------
    print(f"\n=== {model_label} CLASSIFICATION REPORT ===")
    print(classification_report(y_true, y_pred, target_names=["Negative", "Positive"]))

    # -------------------------
    # CONFUSION MATRIX
    # -------------------------
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Negative", "Positive"])
    disp.plot(cmap="Blues" if model_label == "BERT" else "Greens")
    plt.title(f"Confusion Matrix - {model_label}")
    plt.show()

    # -------------------------
    # ROC CURVE
    # -------------------------
    fpr, tpr, _ = roc_curve(y_true, y_prob)
    roc_auc_curve = auc(fpr, tpr)

    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, label=f"{model_label} ROC Curve (AUC = {roc_auc_curve:.4f})")
    plt.plot([0, 1], [0, 1], "--")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"ROC Curve - {model_label}")
    plt.legend()
    plt.grid(True)
    plt.show()

    # -------------------------
    # TRAIN / VAL LOSS CURVE
    # -------------------------
    plot_loss_curves(trainer, model_label)

    # -------------------------
    # SAVE DETAILED RESULTS
    # -------------------------
    token_stats_df = pd.DataFrame([train_token_stats, val_token_stats, test_token_stats])
    token_stats_path = os.path.join(output_dir, f"{model_label.lower()}_token_stats.csv")
    token_stats_df.to_csv(token_stats_path, index=False)

    detailed_results = {
        "Model": model_label,
        "Accuracy": round(metrics["accuracy"], 4),
        "Precision": round(metrics["precision"], 4),
        "Recall": round(metrics["recall"], 4),
        "F1": round(metrics["f1"], 4),
        "ROC-AUC": round(metrics["roc_auc"], 4),
        "Training Time (sec)": round(train_time, 2),
        "Training Time (min)": round(train_time / 60, 2),
        "Inference Time (sec)": round(infer_time, 2),
        "Inference Latency per Sample (ms)": round(inference_latency_per_sample_ms, 4),
        "Training Throughput (samples/sec)": round(train_samples_per_sec, 2),
        "Training Throughput (tokens/sec)": round(train_tokens_per_sec, 2),
        "Inference Throughput (samples/sec)": round(inference_samples_per_sec, 2),
        "Inference Throughput (tokens/sec)": round(inference_tokens_per_sec, 2),
        "Peak GPU Memory Allocated Train (GB)": train_mem_stats["peak_gpu_memory_allocated_gb"],
        "Peak GPU Memory Reserved Train (GB)": train_mem_stats["peak_gpu_memory_reserved_gb"],
        "Peak GPU Memory Allocated Infer (GB)": infer_mem_stats["peak_gpu_memory_allocated_gb"],
        "Peak GPU Memory Reserved Infer (GB)": infer_mem_stats["peak_gpu_memory_reserved_gb"],
        "Device": device_name,
        "Train Mean Tokens": train_token_stats["truncated_mean_tokens"],
        "Val Mean Tokens": val_token_stats["truncated_mean_tokens"],
        "Test Mean Tokens": test_token_stats["truncated_mean_tokens"],
        "Train Truncation %": train_token_stats["truncation_percentage"],
        "Val Truncation %": val_token_stats["truncation_percentage"],
        "Test Truncation %": test_token_stats["truncation_percentage"]
    }

    return {
        "model_label": model_label,
        "trainer": trainer,
        "tokenizer": tokenizer,
        "y_true": y_true,
        "y_pred": y_pred,
        "y_prob": y_prob,
        "metrics": metrics,
        "results_row": detailed_results,
        "token_stats_df": token_stats_df
    }

# =====================================================
# EXPERIMENT SETTINGS
# =====================================================
MAX_LENGTH = 256
BATCH_SIZE = 8
EPOCHS = 2
LEARNING_RATE = 2e-5
WEIGHT_DECAY = 0.01

# =====================================================
# BERT
# =====================================================
bert_results = run_experiment(
    model_label="BERT",
    model_name="bert-base-uncased",
    train_texts=train_texts,
    val_texts=val_texts,
    test_texts=test_texts,
    train_labels=train_labels,
    val_labels=val_labels,
    test_labels=test_labels,
    max_length=MAX_LENGTH,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    learning_rate=LEARNING_RATE,
    weight_decay=WEIGHT_DECAY
)

# =====================================================
# XLNET
# =====================================================
xlnet_results = run_experiment(
    model_label="XLNet",
    model_name="xlnet-base-cased",
    train_texts=train_texts,
    val_texts=val_texts,
    test_texts=test_texts,
    train_labels=train_labels,
    val_labels=val_labels,
    test_labels=test_labels,
    max_length=MAX_LENGTH,
    batch_size=BATCH_SIZE,
    epochs=EPOCHS,
    learning_rate=LEARNING_RATE,
    weight_decay=WEIGHT_DECAY
)

# =====================================================
# FINAL COMPARISON TABLE
# =====================================================
final_results = pd.DataFrame([
    bert_results["results_row"],
    xlnet_results["results_row"]
])

print("\n=== FINAL MODEL COMPARISON ===")
display(final_results)

final_results_path = os.path.join(BASE_OUTPUT_DIR, "fulltext_model_results_with_efficiency.csv")
final_results.to_csv(final_results_path, index=False)

print("\nSaved final comparison to:")
print(final_results_path)

# -------------------------
# COMBINED TOKEN STATS
# -------------------------
combined_token_stats = pd.concat([
    bert_results["token_stats_df"].assign(Model="BERT"),
    xlnet_results["token_stats_df"].assign(Model="XLNet")
], ignore_index=True)

combined_token_stats_path = os.path.join(BASE_OUTPUT_DIR, "combined_token_truncation_stats.csv")
combined_token_stats.to_csv(combined_token_stats_path, index=False)

print("\nSaved token/truncation stats to:")
print(combined_token_stats_path)
```
